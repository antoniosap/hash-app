```python
## https://fullstackml.com/wavelet-image-hash-in-python-3504fdd282b5

## in questo caso l'hash è una ricerca per vicinanza
## le due immagini differiscono per pochissimi pixel. quasi invisibili

## quindi hash1 e hash2 sono uguali per immagini uguali
## nel secondo caso invece l'hash è usato come firma per l'immagine
```

```python
import PIL
from PIL import Image
import imagehash
from IPython.display import display

hash1 = imagehash.phash(Image.open('comics.jpg'))
print(hash1)
pil_im = Image.open('comics.jpg')
display(pil_im)
```

```python
hash2 = imagehash.phash(Image.open('comics-fake.jpg'))
print(hash2)
pil_im2 = Image.open('comics-fake.jpg')
display(pil_im2)
```

```python
## questa è la somiglianza delle due immagini
print(hash1 - hash2)
```

```python
## ma in questo caso l'hash è una firma
from hashlib import sha512
sign1 = sha512(open('comics.jpg', 'rb').read()).hexdigest() 
sign2 = sha512(open('comics-fake.jpg', 'rb').read()).hexdigest()
print(sign1)
print(sign2)
```
